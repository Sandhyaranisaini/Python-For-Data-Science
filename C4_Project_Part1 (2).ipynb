```python
#import libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read file
Car_data = pd.read_excel("/content/DS1_C4_S9_Car_Data.xlsx")
Car_data
```

```python
#Task1:
print(Car_data.head())

# Display the column names
print(Car_data.columns)

# Count occurrences of each body type
body_type_counts = Car_data['Body_Type'].value_counts()

# Get the top 5 most preferred body types
top_5_body_types = body_type_counts.head(5)

# Plot the results using a horizontal bar chart
plt.figure(figsize=(10, 6))
top_5_body_types.plot(kind='barh', color='pink')
plt.xlabel('Number of Cars')
plt.ylabel('Body Type')
plt.title('Top 5 Most Preferred Car Body Types')
plt.gca().invert_yaxis()  # To display the highest count at the top
plt.show()

#Interpretation:here many car model cars are used SUV body type.SUV,Sedan,Hatchback,Coupe,MUV these are the top 5 Car manufacturer body types
```

```python
#task2
# Create a list of required modes
required_modes = {'Normal', 'Comfort', 'Eco', 'Sport', 'Power'}

# Function to check if all required modes are present
def has_all_modes(modes):
    if pd.isna(modes):
        return False
    modes_set = set(modes.split(', '))  # Adjust the split delimiter as needed
    return required_modes.issubset(modes_set)

# Apply the function to filter cars with all required modes
df_with_all_modes = Car_data[Car_data['Drive_Modes'].apply(has_all_modes)]

# Count the number of such cars
count_cars_with_all_modes = df_with_all_modes.shape[0]

print(f"Number of cars with all driving modes (Normal, Comfort, Eco, Sport, Power): {count_cars_with_all_modes}")
#Interpretation:here no cars with all driving modes(Normal, Comfort, Eco, Sport, Power)
```

```python
#Task3:
# Convert the city mileage column to numeric (handle any non-numeric values or errors)
Car_data['City_Mileage_km_litre'] = pd.to_numeric(Car_data['City_Mileage_km_litre'], errors='coerce')

# Group by manufacturer and calculate the mean city mileage
manufacturer_mileage = Car_data.groupby('Make')['City_Mileage_km_litre'].mean()

# Get the top 10 manufacturers based on average city mileage
top_10_manufacturers = manufacturer_mileage.nlargest(10)

# Display the results
print(top_10_manufacturers)

# Plot the top 10 manufacturers using a horizontal bar chart
plt.figure(figsize=(12, 8))
top_10_manufacturers.plot(kind='barh', color='skyblue')
plt.xlabel('Average City_Mileage_km_litre')
plt.ylabel('Manufacturer')
plt.title('Top 10 Car Manufacturers Based on City Mileage')
plt.gca().invert_yaxis()  # To display the highest mileage at the top
plt.show()

# Check for outliers in city mileage
mileage_stats = Car_data['City_Mileage_km_litre'].describe()
print(mileage_stats)

# Plot the distribution of city mileage
plt.figure(figsize=(10, 6))
plt.hist(Car_data['City_Mileage_km_litre'].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.xlabel('City_Mileage_km_litre')
plt.ylabel('Frequency')
plt.title('Distribution of City Mileage')
plt.show()
#Interpretation:Maruti Suzuki,Datsun,Renault,Mahindra,Tata,Lexus,Honda,Hyundai,Volkswagen,Fiat these are the top 10 car manufacturers based on city mileage...Maruti Suzuki is the best car manufacturer based on city mileage
```

```python
#task3(continution):
# Removing outliers using IQR method
Q1 = Car_data['City_Mileage_km_litre'].quantile(0.25)
Q3 = Car_data['City_Mileage_km_litre'].quantile(0.75)
IQR = Q3 - Q1

# Define acceptable range for mileage
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
filtered_df = Car_data[(Car_data['City_Mileage_km_litre'] >= lower_bound) & (Car_data['City_Mileage_km_litre'] <= upper_bound)]

# Re-calculate and plot the top 10 manufacturers
manufacturer_mileage_filtered = filtered_df.groupby('Make')['City_Mileage_km_litre'].mean()
top_10_manufacturers_filtered = manufacturer_mileage_filtered.nlargest(10)

# Plot the results
plt.figure(figsize=(12, 8))
top_10_manufacturers_filtered.plot(kind='barh', color='skyblue')
plt.xlabel('Average City_Mileage_km_litre')
plt.ylabel('Manufacturer')
plt.title('Top 10 Car Manufacturers Based on City Mileage (Filtered)')
plt.gca().invert_yaxis()
plt.show()
```

```python
#Task12:
# Convert necessary columns to numeric
Car_data['Displacement'] = pd.to_numeric(Car_data['Displacement'], errors='coerce')
Car_data['Cylinders'] = pd.to_numeric(Car_data['Cylinders'], errors='coerce')
Car_data['Valves_Per_Cylinder'] = pd.to_numeric(Car_data['Valves_Per_Cylinder'], errors='coerce')

# Function to calculate coefficient of variation
def coefficient_of_variation(series):
    mean = series.mean()
    std_dev = series.std()
    return std_dev / mean if mean != 0 else float('nan')

# Group by manufacturer and calculate statistics
grouped = Car_data.groupby('Make').agg({
    'Displacement': ['mean', 'median', 'std'],
    'Cylinders': ['mean', 'median', 'std'],
    'Valves_Per_Cylinder': ['mean', 'median', 'std']
})

# Calculate coefficient of variation
grouped['Displacement', 'cv'] = Car_data.groupby('Make')['Displacement'].apply(coefficient_of_variation)
grouped['Cylinders', 'cv'] = Car_data.groupby('Make')['Cylinders'].apply(coefficient_of_variation)
grouped['Valves_Per_Cylinder', 'cv'] = Car_data.groupby('Make')['Valves_Per_Cylinder'].apply(coefficient_of_variation)

# Flatten MultiIndex columns
grouped.columns = ['_'.join(col).strip() for col in grouped.columns.values]
grouped = grouped.reset_index()

# Display the statistics
print(grouped)

# Plotting the box plots for Displacement
plt.figure(figsize=(14, 8))
sns.boxplot(data=Car_data, x='Make', y='Displacement')
plt.xticks(rotation=45)
plt.title('Box Plot of Displacement by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Displacement')
plt.grid(True)
plt.show()
#Interpretation:The median displacement (the horizontal line within each box) varies significantly across different manufacturers.
#Some manufacturers have a higher median displacement, indicating that their cars tend to have larger engines.
# The size of the boxes (IQR) also varies, suggesting different levels of variability in displacement among manufacturers.
# Manufacturers with larger boxes have a wider range of displacements within their models
#There are a few outliers visible in the plot, represented by individual points outside the whiskers.
#These outliers might correspond to specific models with significantly larger or smaller displacements compared to the rest of their manufacturer's lineup
```
