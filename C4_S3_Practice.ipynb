```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/DS1_C4_S3_SalesData_Practice.csv")
data
```

```python
#Task1:
sort = data.sort_values("Sales", ascending = True)
sort
```

```python
#TAsk2:
sales= data[(data.Sales >= 45000)]
sales
```

```python
#Task3:
#California
C = data[(data.State == 'California')]                         # Find sales entries for California
print("Total sales in California is:",C['Sales'])
california_sales = C.sort_values('Sales')
california_sales
```

```python
#Texas
T = data[(data.State == 'Texas')]                         # Find sales entries for California
print("Total sales in Texas is:",T['Sales'])
Texas_sales = T.sort_values('Sales')
Texas_sales
```

```python
#Montana
M = data[(data.State == 'Montana')]                         # Find sales entries for California
print("Total sales in Montana is:",M['Sales'])
Montana_sales = M.sort_values("Sales")
Montana_sales
```

```python
#Task4:
print("Total sales in the state of California is:",C['Sales'].sum())
print("Total sales in the state of Texas is:",T['Sales'].sum())
print("Total sales in the state of Montana is:",M['Sales'].sum())
```

```python
#TAsk5:
monthly_sales = california_sales.groupby('Month')['Sales'].sum().reset_index() #aggregation
highest_sales_month = monthly_sales.loc[monthly_sales['Sales'].idxmax()] #find maxsales
lowest_sales_month = monthly_sales.loc[monthly_sales['Sales'].idxmin()] #find minsales

print(f"In the {highest_sales_month['Month']} month the total_sales in California is: {highest_sales_month['Sales']}")
print(f"this is the highest sale for month wise sales")

print(f"In the {lowest_sales_month['Month']} month the total_sales in California is: {lowest_sales_month['Sales']}")
print(f"this is the lowest sale for month wise sales")
```
